```python
import numpy as np
import pandas as pd
import geopandas as gpd
import altair as alt
```

```python
masonry_bldgs = gpd.read_file('data/Unreinforced_Masonry_Buildings_(URM).geojson')
masonry_bldgs['X'] = masonry_bldgs.geometry.x
masonry_bldgs['Y'] = masonry_bldgs.geometry.y
```

```python
masonry_bldgs.shape
```

```python
masonry_bldgs.describe()
```

```python
seattle = gpd.read_file('data/2016_seattle_neighborhoods.geojson') #, driver = 'GeoJSON')
seattle_basemap = alt.Chart(seattle).mark_geoshape(fill='lightgrey', stroke='white', strokeWidth=0.7)

vulnerability_class_order = [None, 'Medium', 'High', 'Critical']

bldg_points = alt.Chart(masonry_bldgs).mark_point(shape = 'triangle', filled = True, color = 'firebrick', size = 15).encode(
    latitude = alt.Latitude('Y'),
    longitude = alt.Longitude('X'),
    color = alt.Color('VULNERABILITY_CLASSIFICATION:N')
)

seattle_bldg_map = seattle_basemap + bldg_points

seattle_bldg_map.project('mercator').properties(
    width = 600,
    height = 600
)
```
